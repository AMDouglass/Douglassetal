```python
from pathlib import Path
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_dir = Path('../Data/Figure4')

fn_list = {i:item for i, item in enumerate(data_dir.rglob('*')) 
           if (not item.is_dir()) & (item.suffix in ['.pickle','.csv','.txt','.xlsx'])}
fn_list
```

```python
path = fn_list[1]
print(path)

df = pd.read_csv(path, skiprows=0, index_col=0)
df.columns = pd.MultiIndex.from_product([['Ad libitum 1','Gastric feeding','Ad libitum 2'], ['Mouse1', 'Mouse2', 'Mouse3', 'Mouse4']])
        
df_melt = df.loc[['AgRP (Zscore)'],:].melt(var_name=['Condition','Mouse'],value_name='Value')
plt.figure()
ax = sns.stripplot(data=df_melt, x="Condition", y="Value", hue='Mouse', size=10, alpha=0.65)
ax.set(ylabel='AgRP (Zscore)')

df_melt = df.loc[['Tb (oC)'],:].melt(var_name=['Condition','Mouse'],value_name='Value')
plt.figure()
sns.stripplot(data=df_melt, x="Condition", y="Value", hue='Mouse', size=10)
ax.set(ylabel='Tb (oC)')

df
```
