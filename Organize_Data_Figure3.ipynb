```python
from pathlib import Path
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_dir = Path('../Data/Figure3')

fn_list = {i:item for i, item in enumerate(data_dir.rglob('*')) 
           if (not item.is_dir()) & (item.suffix in ['.pickle','.csv','.txt','.xlsx'])}
fn_list
```

```python
path = fn_list[0]
print(path)

df = pd.read_csv(path, skiprows=0, index_col=0, usecols=[0,1,2,3])
df.dropna(inplace=True, axis=0)
df.columns = pd.MultiIndex.from_product([['Z-score'], ['Mouse1', 'Mouse2', 'Mouse3']])
                                         
df.plot(y='Z-score', ylabel='Z-score', title=path)
df
```
