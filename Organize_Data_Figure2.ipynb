```python
from pathlib import Path
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def downcast(df, drop=[]):
    ## downcasting loop
    for column in df:
        if df[column].dtype == 'float64':
            df[column]=pd.to_numeric(df[column], downcast='float')
        if df[column].dtype == 'int64':
            df[column]=pd.to_numeric(all_data[column], downcast='integer')
        
    ## dropping an unused columns
    if len(drop):
        df = df.drop(drop, axis=1)
    
    return df

save_dir = Path('../Data/Data/Figure2')
save_dir.mkdir(parents=True, exist_ok=True)
```

```python
data_dir = Path('../Data/Figure2')

fn_list = {i:item for i, item in enumerate(data_dir.rglob('*')) 
           if (not item.is_dir()) & (item.suffix in ['.pickle','.csv','.txt','.xlsx'])}
fn_list
```

```python
path = fn_list[0]
print(path)

df = pd.read_csv(path, skiprows=0)
df.columns = ['Mouse', 'DF/F0', 'Sliding window', 'Food', 'Tb', 'LMA']
df.dropna(inplace=True)

df.mean(axis=1)
sns.stripplot(data=df.drop('Mouse', axis=1).melt(), x="variable", y="value", s=10, alpha=0.65, palette='plasma')
```

```python
df.info(memory_usage = "deep")

downcast(df, drop=[])
df.to_csv(Path(f"{save_dir}/{path.name}"))

df.info(memory_usage = "deep")
```
