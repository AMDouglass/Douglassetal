```python
from pathlib import Path
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def downcast(df, drop=[]):
    ## downcasting loop
    for column in df:
        if df[column].dtype == 'float64':
            df[column]=pd.to_numeric(df[column], downcast='float')
        if df[column].dtype == 'int64':
            df[column]=pd.to_numeric(all_data[column], downcast='integer')
        
    ## dropping an unused columns
    if len(drop):
        df = df.drop(drop, axis=1)
    
    return df

save_dir = Path('../Data/Data/Figure1')
save_dir.mkdir(parents=True, exist_ok=True)
```

```python
data_dir = Path('../Data/Figure1')

fn_list = {i:item for i, item in enumerate(data_dir.rglob('*')) 
           if (not item.is_dir()) & (item.suffix in ['.pickle','.csv','.txt','.xlsx'])}
fn_list
```

### Fig1I

```python
path = fn_list[0]
print(path)

df = pd.read_csv(path, skiprows=2)
df = df.pivot_table(index=['Time'])
df.columns = pd.MultiIndex.from_product([['Z-score'], ['Mouse1', 'Mouse2', 'Mouse3', 'Mouse4']])
                                         
df.plot(y='Z-score', ylabel='Z-score', title=path)
# df
```

```python
df.info(memory_usage = "deep")

downcast(df, drop=[])
df.to_csv(Path(f"{save_dir}/{path.name}"))

df.info(memory_usage = "deep")
```

```python
path = fn_list[7]
print(path)

df = pd.read_pickle(path)
df.plot(x='Time', y='Zscore', title=path)
print(len(df))
# df
```

```python
df.info(memory_usage = "deep")

downcast(df, drop=[])
df['Day'] = df['Day'].astype(np.uint8)
df['ZT'] = df['ZT'].astype(np.uint8)
# df.to_csv(Path(f"{save_dir}/{path.name}"))
# df.info(memory_usage = "deep")

for day, table in df.groupby('Day'):
    df_day = table.drop('Day', axis=1)
    Path(f"{save_dir}/Fig1O").mkdir(parents=True, exist_ok=True)
    df_day.to_pickle(Path(f"{save_dir}/Fig1O/{path.stem}_day-{day}.pickle"))
    print(df_day.info(memory_usage = "deep"))
```
