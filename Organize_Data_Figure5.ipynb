```python
from pathlib import Path
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_dir = Path('../Data/Figure5')

fn_list = {i:item for i, item in enumerate(data_dir.rglob('*')) 
           if (not item.is_dir()) & (item.suffix in ['.pickle','.csv','.txt','.xlsx'])}
fn_list
```

```python
path = fn_list[2]
print(path)

df = pd.read_csv(path, skiprows=0)
# df.columns = ['Mouse', 'DF/F0', 'Sliding window', 'Food', 'Tb', 'LMA']
# df.dropna(inplace=True)

# df.mean(axis=1)
sns.stripplot(data=df.melt(), x="variable", y="value", s=10, alpha=0.65, palette='plasma')
df
```

```python
path = fn_list[6]
print(path)

df = pd.read_pickle(path)
df['TTL'] = df['TTL'].astype(int)
fig, ax = plt.subplots(1,1)
df.plot(x='Time', y='TTL', ax=ax, title=path, color='tomato')
df.plot(x='Time', y='Zscore', ax=ax.twinx(), color='k')
print(len(df))
df
```
